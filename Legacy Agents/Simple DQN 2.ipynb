```python
import random
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook as tqdm
```

```python
# Setup required for Merton Model

# annual calibration
mu = 0.09
sigma = 0.16
rf = 0.01

# static one month investment
dt      = 1/12
horizon = 1/12 
n_steps = int(horizon / dt)

# discretised investment space
Act_Space = np.linspace(-4,6,11)

# return generator
def GBM_Return ():
    # Generate returns from a log normal distribution
    Mean = (mu - (sigma ** 2) / 2) * dt
    Std = sigma * (dt ** 0.5)
    Return = np.random.normal(Mean, Std, 1)[0]

    return np.exp(Return)
```

```python
class NN():
    def __init__(self, Activation = 'Sigmoid', layers = [20, 10]):
        self._Activation_Method = Activation
        self._Weights = None
        self._Weight_Shape = layers
        
    def fit(self, X, Y, learn_rate = 0.001, Loss = 'Onehot', Epoch = 50, **kwarg):
        if isinstance(X, list):
            Input_shape = len(X[0])
            X = np.array(X)
        elif isinstance(X, np.ndarray):
            Input_shape = X.shape[1]
        if isinstance(Y, list):
            Y = np.array(Y).reshape(-1,1)
            
        if self._Weights is None:
            ##Create the weights by initialising them as random values
            previous = Input_shape
            for Layer_Size in self._Weight_Shape:
                if self._Weights is None:
                    self._Weights = [np.random.normal(size = (previous, Layer_Size))/(previous + Layer_Size)**0.5]
                else:
                    Current_Layer = np.random.normal(size = (previous, Layer_Size))/(previous + Layer_Size)**0.5
                    self._Weights.append(Current_Layer)
                previous = Layer_Size
        else:
            pass
        
        for i in range(Epoch):
            ##Forward Propogation and record useful informations
            H_Collect = [X]
            A_Collect = []
            Derivative_Collect = []
            for idx, Weights in enumerate(self._Weights):
                if idx != len(self._Weights)-1:
                    Layer_Active = self._Activation_Method
                else:
                    Layer_Active = 'Identity'

                A = np.matmul(H_Collect[idx], Weights)
                H, derivative = self._Activation(A, how = Layer_Active)
                H_Collect.append(H)
                A_Collect.append(A)
                Derivative_Collect.append(derivative)

            ##Backward Propogation to calculate updated weights
            dW_Collect = []
            dA_Collect = []
            dH_Collect = []
            for idx, Hidden in enumerate(H_Collect[::-1]):
                if idx==0:
                    D_Yhat = self._Loss_Transform(Method = Loss, Y=Y, Y_hat= Hidden, **kwarg)
                    dH_Collect.append(D_Yhat)
                    continue
                else:
                    dA = dH_Collect[-1] * Derivative_Collect[::-1][idx-1]
                    dW = np.matmul(Hidden.transpose(), dA)
                    dH = np.matmul(dA, self._Weights[::-1][idx-1].transpose())
                    dA_Collect.append(dA)
                    dW_Collect.append(dW)
                    dH_Collect.append(dH)

            for i in range(len(self._Weights)):
                self._Weights[i] -= learn_rate * dW_Collect[::-1][i]
            
                    
    def predict(self, X):
        if self._Weights is None:
            print('Please fit your model before you use them to predict')
        else:
            Predictions = X
            for idx, Weights in enumerate(self._Weights):
                if idx != len(self._Weights) - 1:
                    Layer_Active = self._Activation_Method
                else:
                    Layer_Active = 'Identity'
                Predictions = np.matmul(Predictions, Weights)
                Predictions, _ = self._Activation(Predictions, how = Layer_Active)
        return Predictions
                
    def _Activation(self,x, how):
        if how == 'Sigmoid':
            outcome = np.exp(x)/(1 + np.exp(x))
            derivative = outcome*(1-outcome)
        elif how == 'Relu':
            outcome = x*(x>0)
            derivative = np.sign(outcome) 
        elif how == 'Identity':
            outcome = x
            derivative = 1
        elif how == 'Tanh':
            x = np.clip(x, -10, 10)
            outcome = 2/(1+np.exp(-2*x)) - 1
            derivative = 1- outcome**2
        elif how == 'Softplus':
            outcome = np.log(1+np.exp(x))
            derivative = 1/(1+np.exp(-x))
        return outcome, derivative

    def _Loss_Transform(self, Y, Y_hat, Method = 'Onehot', Z = None):
        if Method == 'Onehot':
            #Check Z is in correct form and prepare it for Back Prop
            if Z is None:
                print('For One hot loss method you need to provide Z!!')
            elif isinstance(Z, np.ndarray) == False:
                try:
                    Z = np.array(Z)
                except:
                    print('Z is in an unaccepted form, please Check')        
            if Z.shape[0] != Y.shape[0]:
                Z = Z.transpose()
            #Calculate the Last Layer Back Prop
            Loss = np.zeros(shape = Z.shape)
            for i in range(Z.shape[0]):
                Loss[i, :] = Y[i] * Z[i, :]
            Loss = 2 * (Y_hat*Z - Loss)
        
        if Method == 'Ordinary':
            Loss = 2*(Y_hat - Y)
        return Loss
```

```python
# Agent 1.1 - Fake State Variable
Layers = [5,5]
Layers.append(Act_Space.shape[0])
DQNetwork = NN(Activation= 'Relu', layers= Layers)
batch_size = 512

# Episode Settings
n_episodes    = int(1e4)*3
epsilon_range = [1, 0.01]
N             = n_episodes * n_steps
anneal        = n_episodes / 10
s_dim         = 1

# Return the optimal action index for a single state
def ArgMax_Act(s0, Network = DQNetwork):
    if Network._Weights is None:
        return np.random.choice(Act_Space, 1).astype(int)[0]
    action_qs = Network.predict(s0)
    return np.argmax(action_qs)

# pre-assign exploration for speed
uniforms = np.random.uniform(0, 1, size = N)
epsilons = epsilon_range[0]*np.ones(N) - (1/anneal)*np.arange(0,N)*(epsilon_range[0]-epsilon_range[1])
epsilons = np.maximum(epsilons, epsilon_range[1])
explore = uniforms < epsilons
a_explore = np.random.choice(range(len(Act_Space)), size=N)
plt.figure(); plt.plot(range(N),epsilons); plt.title('Epsilon'); l=plt.xlabel('steps')

# Train up the DQN
exp = []
nans = np.zeros_like(Act_Space)
for i in tqdm(range(n_episodes)):
    s0 = np.ones((1,1))   # Set initial Wealth
    
    for t in range(n_steps):
        idx = a_explore[i] if explore[i] else ArgMax_Act(s0)
        leverage = Act_Space[idx]
        s1 = s0 * (leverage * GBM_Return() + (1 - leverage) * (1 + rf * dt))

        if s1<0:
            nans[idx] += 1 # count failures of discrete time approximation for each a
            break
        else:
            exp.append({'s0' : np.random.uniform(), 'a': idx, 'r': np.log(s1), 's1': s1})
            s0 = s1
        
        # TD Learning - Network is updated after each step.
        # Currently the network itself doesn't conduct SGD, however this
        # mechanic will approximate it.
        if len(exp) > batch_size * 4:
            data = random.sample(exp, batch_size)
            X = np.array([d['s0'] for d in data]).reshape(-1, s_dim)
            Y = [d['r'] for d in data]
            Z = np.zeros((len(Y), len(Act_Space)))

            for k in range(len(data)):
                Z[k, data[k]['a']] = 1
            
            DQNetwork.fit(X, Y,Epoch =1,learn_rate = 0.0008, Z=Z)
```

```python
# Plot the learned value function and optimal Merton policy
print("Nans:", nans, "\n")

f, ax = plt.subplots(1,4, figsize = (16, 4))
for i, s in enumerate(np.random.uniform(size = 4)):
    action_qs = DQNetwork.predict(np.ones((1,1)) * s)
    
    ax[i].plot(Act_Space.reshape((1,-1)), action_qs, '*', color = 'g')
    ax[i].set_title('Q(%.2f,a)' % s)    
    ax[i].set_xlabel('Leverage')
    ax[i].axvline((mu - rf) / (sigma**2), color = 'r')
```
