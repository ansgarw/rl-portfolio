```python
import time
import gym
import matplotlib.pyplot as plt

# Change this line to change DQN version.
import DQN_Agent_IV as DQN_Lib
```

```python
FileName = "Save_FileName.pkl"
```

```python
myEnv = gym.make("MountainCarContinuous-v0")
```

```python
# Asyncronous DQN (Version V)
DQN = DQN_Lib.DQN_Agent("MountainCarContinuous-v0", (8,8,8), Gamma = 0.999, 
                        Epoch = 50, Min_Exp = 5e5, Min_Pos_Exp = 1e4,
                        Over_Sample = True, Replay_Size = 50000,
                        Learning_Rate = 0.01, Batch_Size = 2000, Activation = 'logistic')

# DQN.Load(FileName)
DQN.Initialise_Experiance()
```

```python
# Non-Asyncronous DQN (Version IV or MultiHead)
DQN = DQN_Lib.DQN_Agent(myEnv, (8,8,8), Gamma = 0.999, 
                        Epoch = 50, Min_Exp = 5e5, Min_Pos_Exp = 1e4,
                        Over_Sample = True, Replay_Size = 50000,
                        Learning_Rate = 0.01, Batch_Size = 2000, Activation = 'logistic')

# DQN.Load(FileName)
DQN.Initialise_Experiance()
```

```python
# Perform a Train and plot the results.
DQN.Initialise_Experiance()

Start = time.time()
Rewards, Epsilon = DQN.Train(50)
End = time.time()
print("Time Elapsed: " + str(round(End - Start, 4)))

DQN.Print_Action_Space()
DQN.Print_V_Function()

X = list(range(len(Rewards)))
plt.subplot(2, 1, 1)
plt.plot(X, Rewards, '-', lw=2)
plt.ylabel('Reward')
plt.xlabel('Episodes')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(X, Epsilon, '-', lw=2)
plt.ylabel('Epsilon')
plt.xlabel('Episodes')
plt.show()
```

```python
# Save the Model
DQN.Save(FileName)
print("Save Completed!")
```

```python
# Run the environment with render. Will break jupyter so be sure to save first.
Done = False
State = myEnv.reset()
while Done == False:
    Action = DQN.Decide_Action(State)
    State, Reward, Done, Info = myEnv.step(Action)
    myEnv.render()
myEnv.close()
```
